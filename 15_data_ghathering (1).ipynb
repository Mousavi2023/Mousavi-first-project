```python
import numpy as np
#from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.path as mplPath
from scipy import stats
import os
from IPython.display import clear_output
import rasterio as rio
from rasterio.enums import Resampling
from rasterio.profiles import DefaultGTiffProfile

def CreateFolder(directory):
    try :
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory.' + directory)
# CreateFolder('./'+str(name2)+'_down_sampled')


name = ['01_LST','02_NDVI','03_ODIAC','04_pre','05_t2m','06_gpp','07_LAI','08_windsp','09_winddir','10_soil',
        '11_land_use']

path_list = [[]for i in range (len(name))]
co2_list  = sorted([f for f in os.listdir('./co2/') if f.endswith('.txt')])
for i in range (len(name)): 
    path_list[i] = sorted([f for f in os.listdir('./'+str(name[i])) if f.endswith('.tif')])


for j,name2 in enumerate(co2_list):
    print(name2)
    data =np.loadtxt('./co2/'+str(name2),dtype=[('lat','f8'),('lon','f8'),('time','i4'),('co2','f8'),('var','f8')])
    save =np.zeros(len(data),dtype=[('lat','f8'),('lon','f8'),('time','i4'),('co2','f8')]+
                   [(str(i),'f8') for i in range (len(name))])
#                                ('1','f8'),('2','f8'),('3','f8'),
#                               ('4','f8'),('5','f8'),('6','f8'),('7','f8'),('8','f8'),('9','f8'),('10','f8')])
    save['lat'] = data['lat']
    save['lon'] = data['lon']
    save['time']= data['time']
    save['co2'] = data['co2']
    for i in range (len(name)):
        file    = path_list[i][j]
        print(file)
        ds           = rio.open('./'+str(name[i])+'/'+str(file))
        tif          = ds.read(1)
        tif[tif<-50] = np.nan
        row,col = rio.transform.rowcol(ds.transform,data['lon'],data['lat'])
        rc      = np.array([row,col])
        
        save[str(i)] = tif[rc[0],rc[1]]
    np.savetxt('./mix_'+str(name2),save,fmt='   '.join(['%1.2f']*2+['%i']+['%1.2f']+['%1.2f']+['%1.4f']+
                                                             ['%1.1f']+['%1.3f']+['%1.2f']+['%1.4f']+['%1.2f']+['%1.2f']+
                                                             ['%1.1f']+['%1.4f']+['%1.0f']),
              header='lat  long   time   co2   LST   NDVI   ODIAC   pre    t2m     gpp   LAI   windsp   '+
               'wdir  soil  land_use')
```
