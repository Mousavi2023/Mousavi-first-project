```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.path as mplPath
import os
from scipy.stats import skew
from scipy.stats import kurtosis


name      = '01_LST_mid'
path      = "C:/Users/asus/Desktop/00001_05_deg/00000001_mid/01_LST_0.5_mid/month_LST"

save = [[]for i in range (9)]

path_list = os.listdir(str(path))
k         = 0
for file in (path_list):
    if file.endswith('.tif'):
        data         = np.array(Image.open(str(path)+'/'+str(file)))
        k = k+1
        save[0].append(str(file))
        save[1].append(np.mean(data[data>-10]))
        save[2].append(np.max(data[data>-10]))
        save[3].append(np.min(data[data>-10]))
        save[4].append(np.std(data[data>-10]))
        save[5].append(skew(data[data>-10]))
        save[6].append(kurtosis(data[data>-10]))
        save[7].append(np.sum(data[data>-10]))
        save[8].append(len(data[data>-10]))    

save1 = np.zeros((len(save[0])),dtype=[('name','U20'),('mean','f8'),('min','f8'),('max','f8'),('std','f8'),
                                         ('skew','f8'),('kurt','f8'),('sum','f8'),('N','i8')])

save1['name'] = save[0]
save1['mean'] = save[1]
save1['min']  = save[2]
save1['max']  = save[3]
save1['std']  = save[4]
save1['skew'] = save[5]
save1['kurt'] = save[6]
save1['sum']  = save[7]
save1['N']    = save[8]

print(save1)
np.savetxt('./table_'+str(name)+'_.txt',save1,fmt='      '.join(['%s']+['%1.4f']*7+['%i']),
           header='Name          mean       max      min       std      skew     kurtosis      sum      N_sample')
```
