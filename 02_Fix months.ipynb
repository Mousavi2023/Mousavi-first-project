```python
#OCO2

import geopandas as geo
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.path as mplPath
import time

%matplotlib inline
plt.style.use('classic')

#insert the source address: 
data1 = np.loadtxt('./oco-2_middle_east_2015.txt',dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

data2 = np.loadtxt('./oco-2_middle_east_2016.txt',dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

data3 = np.loadtxt('./oco-2_middle_east_2017.txt',dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

data4 = np.loadtxt('./oco-2_middle_east_2018.txt',dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

data5 = np.loadtxt('./oco-2_middle_east_2019.txt',dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

data6 = np.loadtxt('./oco-2_middle_east_2020.txt',dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

x1 = data1['time'].size
x2 = data2['time'].size
x3 = data3['time'].size
x4 = data4['time'].size
x5 = data5['time'].size
x6 = data6['time'].size


x      = x1 + x2 + x3 + x4 + x5 + x6
data_n = np.zeros(x,dtype=[('lat','f8'),('lon','f8'),('time','i'),('co2','f8'),('var','f8')])

data_n['lat'][:x1] = data1['lat']
data_n['lon'][:x1] = data1['lon']
data_n['co2'][:x1] = data1['co2']
data_n['time'][:x1]= data1['time']
data_n['var'][:x1] = data1['var']

data_n['lat'][x1:x2+x1] = data2['lat']
data_n['lon'][x1:x2+x1] = data2['lon']
data_n['co2'][x1:x2+x1] = data2['co2']
data_n['time'][x1:x2+x1]= data2['time']+12
data_n['var'][x1:x2+x1] = data2['var']

data_n['lat'][x2+x1:x3+x2+x1] = data3['lat']
data_n['lon'][x2+x1:x3+x2+x1] = data3['lon']
data_n['co2'][x2+x1:x3+x2+x1] = data3['co2']
data_n['time'][x2+x1:x3+x2+x1]= data3['time']+24
data_n['var'][x2+x1:x3+x2+x1] = data3['var']

data_n['time'][(data_n['time']>31)&(data_n['time']<=35)]=data_n['time'][(data_n['time']>31)&(data_n['time']<=35)]+1

data_n['lat'][x3+x2+x1:x4+x3+x2+x1] = data4['lat']
data_n['lon'][x3+x2+x1:x4+x3+x2+x1] = data4['lon']
data_n['co2'][x3+x2+x1:x4+x3+x2+x1] = data4['co2']
data_n['time'][x3+x2+x1:x4+x3+x2+x1]= data4['time']+36
data_n['var'][x3+x2+x1:x4+x3+x2+x1] = data4['var']

data_n['lat'][x4+x3+x2+x1:x5+x4+x3+x2+x1] = data5['lat']
data_n['lon'][x4+x3+x2+x1:x5+x4+x3+x2+x1] = data5['lon']
data_n['co2'][x4+x3+x2+x1:x5+x4+x3+x2+x1] = data5['co2']
data_n['time'][x4+x3+x2+x1:x5+x4+x3+x2+x1]= data5['time']+48
data_n['var'][x4+x3+x2+x1:x5+x4+x3+x2+x1] = data5['var']


data_n['lat'][x5+x4+x3+x2+x1:] = data6['lat']
data_n['lon'][x5+x4+x3+x2+x1:] = data6['lon']
data_n['co2'][x5+x4+x3+x2+x1:] = data6['co2']
data_n['time'][x5+x4+x3+x2+x1:]= data6['time']+60
data_n['var'][x5+x4+x3+x2+x1:] = data6['var']

np.savetxt('./oco2_co2_buf_middle_east_new.txt',data_n,fmt='     '.join(['%1.4f']*2+['%i']+['%1.4f']+['%1.4f']),
           header= 'latitude  longitud    time      co2    variance')
```
